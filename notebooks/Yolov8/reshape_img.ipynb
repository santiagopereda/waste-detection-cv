```python
import pandas as pd
import numpy as np
import base64
import os
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image_path = './images'
```

```python
def reshape_img(image_path):
    '''
    Input a the path of the image and output its resized
    version (640x640)
    '''
    #Count the number of images contained in the "./image" file and storing th count in num_files
    
    #Getting the image from the ./images directory
    im = os.listdir(image_path)[0]
    
    #Opening the image using the .open method from PIL library
    img = Image.open(image_path+'/'+im)
    
    # resizing the image to 640x640
    img_resized = img.resize((640,640))
    print('Image resized✅')
    return img_resized
        
```

```python
plt.imshow(reshape(image_path))
```
